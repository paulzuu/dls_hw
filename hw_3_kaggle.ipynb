```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
```

```python
# YOUR CODE
#  ohe  

ohe = OneHotEncoder(sparse_output=False, handle_unknown='error')

data_unprocessed = data[cat_cols]
ohe_data = ohe.fit_transform(data_unprocessed)
ohe_features = pd.DataFrame(ohe_data,
                            columns=ohe.get_feature_names_out(data_unprocessed.columns),
                            index=data_unprocessed.index
                            )

data_processed = pd.concat([data, ohe_features], axis=1)
data_processed = data_processed.drop(data_unprocessed, axis=1)
```

```python
data
```

```python
data_processed
```

```python
y = data_processed['Churn']
X = data_processed.drop(['Churn'], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=True, test_size=0.2, random_state=42)
```

```python
#   

scaler = StandardScaler()

X_train_scale = scaler.fit_transform(X_train)
X_test_scale = scaler.transform(X_test)
```

```python
Cs = [100, 10, 1, 0.1, 0.01, 0.001]

lr_cv = LogisticRegressionCV(Cs=Cs, cv=5, penalty='l2', scoring='roc_auc', solver='lbfgs', refit=True)

lr_cv.fit(X_train_scale, y_train)


y_train_pred = lr_cv.predict_proba(X_train_scale)[:, 1]
y_test_pred = lr_cv.predict_proba(X_test_scale)[:, 1]
```

```python
print('roc-auc for train: %.4f' % roc_auc_score(y_train, y_train_pred))
print('roc-auc for validation: %.4f' % roc_auc_score(y_test, y_test_pred))
```

```python
#         
y_origin = data_cb['Churn']
X_origin = data_cb.drop(['Churn'], axis=1)
```

```python
X_orig_train, X_orig_test, y_orig_train, y_orig_test = train_test_split(X_origin, y_origin, shuffle=True, test_size=0.2, random_state=42)
```

```python
X_origin
```

```python
!pip install optuna
```

```python
from catboost import CatBoostClassifier, Pool
import optuna

train_pool = Pool(X_orig_train, y_orig_train, cat_features=cat_cols)
test_pool = Pool(X_orig_test, y_orig_test, cat_features=cat_cols)

def search_param(trial):

  params = {'iterations': trial.suggest_int('iterations', 100, 1000),
            'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3),
            'depth': trial.suggest_int('depth', 4, 10),
            'verbose': False,
            'random_state': 42
            }

  cbc = CatBoostClassifier(**params)

  cbc.fit(train_pool)

  y_pred = cbc.predict_proba(test_pool)[:, 1]
  ras = roc_auc_score(y_orig_test, y_pred)
  return ras

study = optuna.create_study(direction='maximize')
study.optimize(search_param, n_trials=50)

print(' :', study.best_params)
print(' ROC-AUC:', study.best_value)

best_model = CatBoostClassifier(**study.best_params, verbose=False, random_state=42)
best_model.fit(train_pool)


y_pred_val_cbc = best_model.predict_proba(test_pool)[:, 1]

print('ROC-AUC  :', roc_auc_score(y_test, y_pred_val_cbc))
```
